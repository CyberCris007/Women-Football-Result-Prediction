```python
import pandas as pd

```

```python
results_csv_link = 'https://raw.githubusercontent.com/bechosen-spec/Women-Football-Result-Prediction/main/results.csv'
results_csv = pd.read_csv(results_csv_link)

results_csv
```

```python
results_csv.dtypes
```

```python
results_csv['date'] = results_csv['date'].astype('datetime64')
results_csv['home_team'] = results_csv['home_team'].astype('string')
results_csv['away_team'] = results_csv['away_team'].astype('string')
results_csv['tournament'] = results_csv['tournament'].astype('string')
results_csv['city'] = results_csv['city'].astype('string')
results_csv['country'] = results_csv['country'].astype('string')
```

```python
results_csv.info()
```

```python
shootouts_csv_link = 'https://raw.githubusercontent.com/bechosen-spec/Women-Football-Result-Prediction/main/shootouts.csv'

shootouts_csv = pd.read_csv(shootouts_csv_link)

shootouts_csv
```

```python
shootouts_csv.dtypes
```

```python
newtype = {'date' : 'datetime64', 'home_team' : 'string', 'away_team' : 'string', 'winner' : 'string'}

shootouts_csv = shootouts_csv.astype(newtype)
```

```python
shootouts_csv.info()
```

```python
results_csv.to_csv('cleaned_results.csv', index=False)

```

```python
shootouts_csv.to_csv('cleaned_shootouts.csv', index=False)
```
